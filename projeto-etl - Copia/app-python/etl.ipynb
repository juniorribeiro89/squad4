```python
# Crie um cursor para executar consultas SQL
cur = conn.cursor()

# Liste todas as tabelas no banco de dados atual
cur.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';")
tables = cur.fetchall()

# Imprima as tabelas
for table in tables:
    print(table[0])

# Feche o cursor e a conexão
cur.close()
conn.close()
```

```python
import psycopg2
import pandas as pd

# Conectar ao PostgreSQL
conn = psycopg2.connect(
    database="my_database_etl",
    user="root",
    password="examplepassword",
    host="localhost",
    port="5432"
)

# Criar um cursor
cur = conn.cursor()

# Executar a consulta SQL
cur.execute("SELECT * FROM products;")

# Recuperar os resultados em um DataFrame
columns = [desc[0] for desc in cur.description]
results = cur.fetchall()
df = pd.DataFrame(results, columns=columns)

# Excluir a coluna 'product_id'
df = df.drop(columns=['product_id, '])

# Renomear as colunas para nomes em português
df = df.rename(columns={
    'product_category_name': 'Nome da Categoria',
    'product_name_length': 'Comprimento do Nome',
    'product_description_length': 'Comprimento da Descrição',
    'product_photos_qty': 'Quantidade de Fotos',
    'product_weight_g': 'Peso do Produto (g)',
    'product_length_cm': 'Comprimento do Produto (cm)',
    'product_height_cm': 'Altura do Produto (cm)',
    'product_width_cm': 'Largura do Produto (cm)'
})

# Visualizar os 10 primeiros registros com as colunas renomeadas
print(df.head(10))

# Fechar o cursor e a conexão
cur.close()
conn.close()
```
